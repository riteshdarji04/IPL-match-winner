```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df_deliveries = pd.read_csv('deliveries.csv',)
df_deliveries.info()
df_deliveries.head(10)
```

```python
df_matches = pd.read_csv('matches.csv',)
df_matches.info()
df_matches.head(50)
```

```python
total_score_df = df_deliveries.groupby(['match_id','inning']).sum()['total_runs'].add(1).reset_index()
```

```python
total_score_df = total_score_df[total_score_df['inning']==1]
total_score_df
```

```python
# only first innning score of all matches
match_df = df_matches.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
match_df
```

```python
match_df['team1'].unique()
```

```python
#data cleaning
teams = [
    'sunrisers hyderabad',
    'mumbai indians',
    'royal challengers bangalore',
    'kolkata knight riders',
    'kings xi punjab',
    'chennai super kings',
    'rajasthan royals',
    'delhi capitals'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','delhi capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','delhi capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','sunrisers hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','sunrisers hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
# matches where DLS method is not appllied




```

```python
match_df = match_df[['id','city','winner','total_runs']]
```

```python
match_df = match_df.rename(columns={'id':'match_id'})
```

```python
delivery_df = match_df.merge(df_deliveries,on='match_id')
```

```python
# for only second inning

delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df
```

```python
# for claculatin how muche balls are lef and how much runs remaeaning after every ball so we hav eto use the cummulative sum on 'total_runs_y'


# Ensure 'total_runs_y' is numeric
delivery_df['total_runs_y'] = pd.to_numeric(delivery_df['total_runs_y'])

# Calculate cumulative sum within each group
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].transform('cumsum')

# delivery_df.drop(columns=['cumulative_runs'], inplace=True)

delivery_df
# delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']


```

```python
# total_runs_x = target runs
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df.head(10)
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df.head(10)
# delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x=="0" else "1")
# delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
# wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed']
# delivery_df['wickets_remaining'] = 10-wickets
```

```python
# Step 2: Replace player names with 1, keep "0" as is
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: 0 if x == "0" else 1)
delivery_df.head(10)
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')

delivery_df.head(10)
```

```python
# Step 4: Calculate the cumulative sum of wickets for each match and inning
delivery_df['cumulative_wickets'] = delivery_df.groupby(['match_id', 'inning'])['player_dismissed'].cumsum()
delivery_df.head(10)
```

```python
# Step 5: Calculate the number of wickets left
delivery_df['wickets_left'] = 10 - delivery_df['cumulative_wickets']
# Display the first few rows to check
delivery_df.head(20)
```

```python
# calculating current run-rate
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
# calculating required run-rate
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
# winner= batting team then 1 otherwise 0

def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
delivery_df
```

```python
# fetching the required data only
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','total_runs_x','crr','rrr','result']]
```

```python
final_df
```

```python
# data can be biased because the data of the same match is together so we have to suffle it

final_df = final_df.sample(final_df.shape[0])
```

```python
# Dropping NUll Values
final_df.dropna(inplace=True)
```

```python
# Replace infinite values in column 'crr' with the large value
final_df['crr'] = final_df['crr'].replace([np.inf, -np.inf], 1e10)
```

```python
final_df.sample()
# in below result Chennai Super Kings is loosing team because result is 0
```

```python
final_df.isnull().sum()
```

```python
# Removing rows where balls_left==0 to remove -inf and inf from rrr column
final_df = final_df[final_df['balls_left']!=0]
```

```python
# Storing data in new csv
final_df.to_csv("final_df.csv", index=False)
```

```python
# building model logistic regression

from sklearn.model_selection import train_test_split as tts
```

```python
x = final_df.iloc[:, :-1]
y = final_df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = tts(x,y,test_size=0.25,random_state=100)
x_train
```

```python
x_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(x_test)[0]
```

```python
# Random Forest classifier


from sklearn.ensemble import RandomForestClassifier
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
],
remainder='passthrough')

pipe = Pipeline([
    ('step1',trf),
    ('step2',RandomForestClassifier())
])

pipe.fit(x_train,y_train)

y_pred = pipe.predict(x_test)

accuracy_score(y_test,y_pred)
```
